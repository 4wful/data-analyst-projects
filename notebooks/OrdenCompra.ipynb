```python
import numpy as np
import pandas as pd
```

```python
#Cargar DataSet
df=pd.read_csv('../data/raw/DATASETS - OC.csv')
#Ver DataFrame
df.head()
```

LIMPIEZA DE DATOS

```python
#Corrección de nombre de las columnas 
df.columns=[
    'fecha_corte',
    'departamento',
    'provincia',
    'distrito',
    'ubigeo',
    'tipo_orden',
    'año',
    'mes',
    'ruc',
    'índice',
    'fecha_orden',
    'monto_soles',
    'razón_social',
    'descripción',
    'estado_orden',
    'expediente_siaf'
]
#verificar cambios
df.head()
```

```python
#Eliminar duplicados ( si es que hay)
# Ver el número de filas antes
print("Antes:", df.shape)

# Eliminar duplicados (por defecto busca duplicados en todas las columnas)
df_sin_duplicados = df.drop_duplicates()

# Ver el número de filas después
print("Después:", df_sin_duplicados.shape)
```

```python
#Correguir datos nulos o faltantes
#Validando tipos de datos
print(df.dtypes)
# Ver valores nulos por columna
print(df.isnull().sum())
```

TRANSFORMACIÓN DE DATOS

```python
#Correccion de formato FECHA
df['fecha_corte'] = pd.to_datetime(df['fecha_corte'], format='%Y%m%d')
df['fecha_orden'] = pd.to_datetime(df['fecha_orden'], format='%Y%m%d')
# Reemplazar el punto decimal por coma en 'monto_soles', formateando a 2 decimales
df['monto_soles'] = df['monto_soles'].apply(lambda x: f"{x:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
#Creación de una columna para identificar el nombre del mes
df['mes_nombre'] = df['fecha_orden'].dt.strftime('%B')  # Para nombre del mes
#verificar cambios
df.head()
```

```python
#Homogenización del texto 
df['departamento'] = df['departamento'].str.title().str.strip()
df['razón_social'] = df['razón_social'].str.upper().str.strip()
df['estado_orden'] = df['estado_orden'].str.strip()
```

```python
#Ahorrando memoria mediante cetegoría
categorical_cols = ['departamento', 'provincia', 'distrito', 'tipo_orden', 'estado_orden', 'mes_nombre']
for col in categorical_cols:
    df[col] = df[col].astype('category')
```

```python
#Guardar DataFrame limpio
df.to_csv('../data/processed/DATASETS - OC_limpio.csv', index=False)
#Verificando el nuevo df final
df.head()
```
