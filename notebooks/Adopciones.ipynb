```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar datos con codificación adecuada
df = pd.read_csv('../data/raw/1.4.1 BdD_Adopciones_5.csv', sep=';', encoding='latin1')

# Visualizar datos
df.head()
```

Limpieza de Datos

```python
#Eliminar duplicados ( si es que hay)
# Ver el número de filas antes
print("Antes:", df.shape)

# Eliminar duplicados (por defecto busca duplicados en todas las columnas)
df_sin_duplicados = df.drop_duplicates()

# Ver el número de filas después
print("Después:", df_sin_duplicados.shape)
```

```python
#Correguir datos nulos o faltantes
# Ver valores nulos por columna
df.isnull().sum()
```

```python
#Los nulos son valores "faltantes" o "vacios" en este caso revisnado el documento son faltantes
#Lo reemplazaremos por el valor "0"
# Reemplazar campos de ubicación vacíos por "0"
ubicaciones = [
    'nna_adoptados_hombres',
    'nna_adoptados_mujeres',
    'nna_adoptados_menores_1_total',
    'nna_adoptados_menores_1_hombres',
    'nna_adoptados_menores_1_mujeres',
    'nna_adoptados_1_5_total',
    'nna_adoptados_1_5_hombres',
    'nna_adoptados_1_5_mujeres',
    'nna_adoptados_6_11_total',
    'nna_adoptados_6_11_hombres',
    'nna_adoptados_6_11_mujeres',
    'nna_adoptados_12_17_total',
    'nna_adoptados_12_17_hombres',
    'nna_adoptados_12_17_mujeres',
    'nna_tipo_regular',
    'nna_tipo_prioritario',
    'familia_peruana',
    'familia_extranjera',
    'proceso_dit',
    'proceso_juzgado',
    'prioritario_salud',
    'prioritario_necesidades_especiales',
    'prioritario_grupo_hermanos',
    'prioritario_mayores_9',
    'prioritario_adolescentes',
    'prioritario_interes_superior',
    'prioritario_familia_acogedora'
]

for col in ubicaciones:
    df[col] = df[col].fillna('NO DECLARADO')
# Ver valores nulos por columna
df.isnull().sum()
```

TRANSFORMACIÓN DE DATOS

```python
# Dividir la columna 'periodo_informacion' en dos columnas usando el guion "-"
df[['periodo_inf_inicio', 'periodo_inf_fin']] = df['periodo_informacion'].str.split('-', expand=True)
#Visualizar los datos correguidos
df.head(30)
```

```python
# Guardar el DataFrame limpio en un archivo CSV
df.to_csv('../data/processed/1.4.1 BdD_Adopciones_5_limpio.csv', index=False)
```

VISUALIZACIÓN PRELIMINAR 

```python
# Contar la frecuencia de los valores únicos en la columna 'prioritario_salud'
frecuencia = df['prioritario_salud'].value_counts()

# Gráfico de barras
plt.figure(figsize=(8, 6))
frecuencia.plot(kind='bar', color='skyblue')

# Títulos y etiquetas
plt.title('Distribución de Prioritario Salud', fontsize=16)
plt.xlabel('Prioritario Salud', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Mostrar el gráfico
plt.show()
```
