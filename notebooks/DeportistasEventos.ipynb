```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar datos con codificación adecuada
df = pd.read_csv('../data/raw/POSTULANTES-UNE.csv', sep=';', encoding='latin1')

# Visualizar datos
df.head()
```

LIMPIEZA DE DATOS

```python
#Corrección Nombres de Columnas para una mejor comprensión
df.columns = [
    'fecha_corte',
    'uuid',
    'facultad',
    'especialidad',
    'modalidad',
    'nivel_academico',
    'año_postulacion',
    'departamento',
    'provincia',
    'distrito',
    'departamento_origen',
    'provincia_origen',
    'distrito_origen',
    'colegio',
    'sexo',
    'discapacidad',
    'etnia'
]
#Visualizar los datos correguidos
df.head()
```

TRANSFORMACIÓN DE DATOS 

```python
#Correccion de formato FECHA
df['fecha_corte'] = pd.to_datetime(df['fecha_corte'], format='%Y%m%d')
# Extraer año y semestre
df[['año', 'semestre']] = df['año_postulacion'].str.split('-', expand=True)
#Visualizar los datos correguidos
df.head()
```

```python
# Ver valores nulos por columna
df.isnull().sum()
```

```python
# Reemplazar campos de ubicación vacíos por "NO DECLARADO"
ubicaciones = ['departamento', 'provincia', 'distrito',
               'departamento_origen', 'provincia_origen', 'distrito_origen']

for col in ubicaciones:
    df[col] = df[col].fillna('NO DECLARADO')

# Reemplazar colegio vacío por "NO REGISTRADO"
df['colegio'] = df['colegio'].fillna('NO REGISTRADO')
```

```python
# Ver valores nulos por columna
df.isnull().sum()
```

```python
#Visualizar datos
df.head()
```

```python
#Preevenir Valores Inconsistentes
df['sexo'] = df['sexo'].str.upper().str.strip()
df['colegio'] = df['colegio'].astype(str).str.strip().str.title()
df['discapacidad'] = df['discapacidad'].str.upper().str.strip()
df['etnia'] = df['etnia'].str.upper().str.strip()
```

```python
#Guardar DataFrame limpio
df.to_csv('../data/processed/POSTULANTES-UNE_limpio.csv', index=False)
```
