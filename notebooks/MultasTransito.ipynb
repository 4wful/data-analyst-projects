```python
import numpy as np
import pandas as pd
```

```python
# Cargar datos con codificación adecuada
df = pd.read_csv('../data/raw/Multas_transito_Ene-Mar_2025.csv', skiprows=2)
# Visualizar datos
df.head()
```

```python
#Borrar columnas innecesarias
df = df.drop(['Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16'], axis=1)
#Corrección Nombres de Columnas para una mejor comprensión
df.columns = [
    'codigo_falta',
    'descripción',
    'tipo_formato',
    'nivel_gravedad',
    'año_infracción',
    'mes_infracción',
    'cant_impuesta',
    'importe_impuesto_soles',
    'reincidencia_impuesta_soles'
]
#Visualizar los datos correguidos
df.head()
```

```python
num_filas_vacias = df[df.isnull().all(axis=1)].shape[0]
print(f"Filas completamente vacías: {num_filas_vacias}")
```

```python
df = df.dropna(how='any')
print(f"Nueva cantidad de filas: {df.shape[0]}")
```

```python
# Ver valores nulos por columna
df.isnull().sum()
```

TRANSFORMACIÓN DE DATOS

```python
# Año y Mes a valor entero en 64 bits
df['año_infracción'] = df['año_infracción'].astype('int64')
df['mes_infracción'] = df['mes_infracción'].astype('int64')
# Reemplazar comas y convertir a int64
df['cant_impuesta'] = df['cant_impuesta'].str.replace(',', '').astype('int64')
df['importe_impuesto_soles'] = df['importe_impuesto_soles'].str.replace(',', '').astype('int64')
df['reincidencia_impuesta_soles'] = df['reincidencia_impuesta_soles'].str.replace(',', '').astype('int64')
# Verificar cambios
df.head()
```

```python
# Guardar el DataFrame limpio en un archivo CSV
df.to_csv('../data/processed/Multas_transito_Ene-Mar_2025_limpio.csv', index=False)
```
